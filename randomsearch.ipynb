```python

import matplotlib.pyplot as plt
import numpy as np
import numpy.random
```

```python
numpy.random.seed(7)
```

```python
def objective(x):
  return x**4-x**2+2*x-8
```

```python
x_min = -10
x_max = 10

x = np.arange(x_min-1,x_max+1,0.1)
sample = x_min + np.random.rand(150)*(x_max-x_min)


y = objective(sample)

opt_ind = 0
for i in range(len(sample)):
  if y[i] < y[opt_ind]:
    opt_ind = i


print(f"X-Optimal : {sample[opt_ind]} , Y-Optimal : {y[opt_ind]}")
plt.figure(figsize = (18,6))
plt.plot(x , objective(x) , color = "green" , label = "y")
plt.scatter(sample , y , color = "blue" , label = "Optimal")
plt.scatter(sample[opt_ind] , y[opt_ind] , color = "red" , label = "Optimal")
plt.legend()
plt.show()


```
